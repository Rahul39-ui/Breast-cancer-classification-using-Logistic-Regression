```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/studygyaan/Cancer-Type-Classification/master/BCA.csv')
```

```python
data.head()
```

```python
col_name=data.columns
```

```python
col_name
```

```python
y=data.diagnosis
drop_cols=['id','Unnamed: 32','diagnosis']
x=data.drop(drop_cols,axis=1)
x.head()
```

```python
ax=sns.countplot(y,label='Count',palette='Blues')
B,M=y.value_counts()
print('Benign tumor is',B)
print('Malignant tumor is',M)
```

```python
x.describe()
```

```python

sns.set(style='whitegrid')
data=x
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,0:10]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='value')
figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis',data=data,palette='Blues')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style='whitegrid')
data=x
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,10:20]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='value')
figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis',data=data,palette='Purples')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style='whitegrid')
data=x
data_std=(data-data.mean())/data.std()
data=pd.concat([y,data_std.iloc[:,20:30]],axis=1)
data=pd.melt(data,id_vars='diagnosis',var_name='features',value_name='value')
figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis',data=data,palette='autumn')
plt.xticks(rotation=45)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(20,20))
sns.heatmap(x.corr(),annot=True,linewidth=0.5,fmt='.1f',ax=ax)
plt.show()
```

```python
x.head()
```

```python
drop_list=['radius_mean', 'perimeter_mean', 'compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst', 'concavity_worst','compactness_se','concave points_se','texture_worst','area_worst']
x1=x.drop(drop_list,axis=1)
x1.head()
```

```python
f,ax=plt.subplots(figsize=(20,20))
sns.heatmap(x1.corr(),annot=True,linewidth=0.5,fmt='.1f',ax=ax)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.2, random_state=21)
clf_rf=RandomForestClassifier(random_state=42)
clf_rf=clf_rf.fit(X_train,y_train)

ac=accuracy_score(y_test,clf_rf.predict(X_test))
print('Accuracy is:',ac*100)
cm=confusion_matrix(y_test,clf_rf.predict(X_test))
sns.heatmap(cm/np.sum(cm),annot=True,fmt='0.2%',cmap='Blues')
plt.show()
```
